```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import os

os.chdir(r"C:\Users\Admin\Desktop\regularization")
```

```python
data=pd.read_csv(r"datadata.txt")
data
```

```python
df=data.copy()
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.linear_model import Ridge
```

```python
l=.3
ridge=Ridge(alpha=l)
ridge.fit(x,y)
```

```python
ridge_coff=ridge.coef_
ridge_coff
```

```python
ridge_intercept=ridge.intercept_
ridge_intercept
```

```python
x_test=[1,2,3,4]
y_pred=ridge.predict(pd.DataFrame(x_test))
y_pred
```

```python
plt.figure(figsize=(12,6),dpi=75)
plt.plot(x_test,y_pred,color="red",linewidth=2)
plt.scatter(x,y)
plt.ylim(ymin=0,ymax=9)
plt.xlim(xmin=0,xmax=7)
plt.text(x_test[-1],y_pred[-1], '  y ='+ str('%.2f' %ridge_coff) +'* x +' + str('%.2f' %ridge_intercept)+  '  For \u03BB or  \u03b1 = ' +str(l) ,fontsize=12)
plt.title('ridge Regression')
plt.show()
```
